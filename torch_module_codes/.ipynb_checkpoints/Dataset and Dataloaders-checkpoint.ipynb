```python
import torch 
if torch.cuda.is_available():
    torch.device('cuda')
import torchvision
from torch.utils.data import Dataset,DataLoader
import numpy as np
import math
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
class RSNE_DATASET(Dataset):
    def __init__(self):
        #data loading
        self.df=pd.read_csv('stage_2_train_labels.csv')
        X=self.df.drop(['Target','patientId'],axis=1).values
        y=self.df['Target'].values
        sc=StandardScaler()
        X_train=sc.fit_transform(X)
        y_train=y
        
        self.X_train=torch.tensor(X_train,dtype=torch.float32)
        self.y_train=torch.tensor(y_train,dtype=torch.float32)
        
    def __getitem__(self,index):
        return self.X_train[index],self.y_train[index]
        
    def __len__(self):
        #return(len(self.df))
        #length of dataset
        return len(self.X_train)
    
    
```

```python
dataset=RSNE_DATASET()
random_data=dataset[200]
features,labels=random_data
print(features,labels)
```

```python
#DATALOADER
dataloader=DataLoader(dataset=dataset,batch_size=4,shuffle=True,num_workers=2)
```

```python
data_iter=iter(dataloader)
data=data_iter.next()
features,labels=data
print(features,labels)
#since batch size is 4 we see 4 vectors
```

```python
#dummy training

num_epochs=2
batch_size=4
total_samples =len(dataset)
n_iterations=math.ceil(total_samples/batch_size)
print(total_samples,n_iterations)



for epoch in range(num_epochs):
    for i,(inputs,labels) in enumerate(dataloader):
        if (i+1)%100==0:
            print(f'epoch:{epoch+1}/{num_epochs}, inputs{inputs.shape}')
        
```
