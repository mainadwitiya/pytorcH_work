### A very simple notebook for various activation functions.

I will keep adding more functions in a few days and also the code for pytorch implementation.

```python
import numpy as np

array1 = np.array([[1,7,3],[4,1,6]])
```

### Sigmoid Activation
Value ranges from 0 to 1


```python
def sigmoid_fuunc(X):
    return 1/(1+np.exp(-X))
result =sigmoid_fuunc(array1)
print(f'Sigmoid Output :: {result}')
```

### ReLU Activation
Vv straigh forward


```python
def relu_func(X):
    return np.maximum(0,X)
result =relu_func(array1)
print(f'ReLu Output :: {result}')
```

### Softmax Activation
The Softmax function produces an output which is a range of values between 0 and 1
Note- Softmax are used because Linear layers digest the vectors but doesn't add to 1.


```python
def softmax(X):
    return np.exp(X)/np.sum(np.exp(X))
result =softmax(array1)
print(f'Softmax Output :: {result}')
```
